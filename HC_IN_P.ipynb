Well, firstly suppose, that I havea collection of data points represented by a numpy array as follows:

```python
import numpy as np

X = np.array([
    [6,4],  
    [11,17],
    [17,13],
    [28,19],
    [40,40],
    [86,74],
    [74,80],
    [50,74],
    [70,55],
    [82,92],
    ])
```

```python
import matplotlib.pyplot as plt

labels = range(1, 11)  
plt.figure(figsize=(10, 7))  
plt.subplots_adjust(bottom=0.1)  
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):  
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show() 
```

Here we can see that we can make different clusters. 

It is also worth noting, that in real life we have thousands of data points and most of the time is impossible to notice the clusters. However because of educational purpose I will try do clustering with less data. 

## Clustering with Single method

```python
from scipy.cluster.hierarchy import dendrogram, linkage  
from matplotlib import pyplot as plt

linked1 = linkage(X, 'single')

labelList = range(1, 11)

plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()  
```

Here we can see that algorith starts by finding the two points that are closest to each other (Euclidean distance)

To 2 and 3 it was added 4 then 1. Also 5 is seprated cluster, however, it is near to 2,3,4,1.
Because 2 and 3 is closer to each other, the Euclidean distance is lower than 6 and 7.
